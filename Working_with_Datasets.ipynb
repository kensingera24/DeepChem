```python
import deepchem as dc
dc.__version__
```

```python
# Load the Delaney dataset of molecular solubilities
tasks, datasets, transformers = dc.molnet.load_delaney(featurizer='GraphConv')
train_dataset, valid_dataset, test_dataset = datasets
```

```python
# Take a look at each of the datasets
print(test_dataset)
```

```python
# Access all data from the dataset
test_dataset.y
```

```python
# Iterate over the dataset to load a little bit of data at a time
# This frees memory before loading the next bit
for X,y,w,id in test_dataset.itersamples():
  print(y,id)
```

```python
# Iterate over a batch of samples
for X,y,w,ids in test_dataset.iterbatches(batch_size=50):
  print(y.shape)
```

```python
# iterbatches() has other features that are useful when training models
# You can iterate over the complete dataset ten times, each time
# with the samples in a different random order

for X,y,w,ids in test_dataset.iterbatches(batch_size=10, epochs=10, deterministic=False):
  print(y.shape)
```

```python
# View the data using a data frame
test_dataset.to_dataframe()
```

```python
# Creating datasets
import numpy as np
X = np.random.random((10,5))
y = np.random.random((10,2))
dataset = dc.data.NumpyDataset(X=X, y=y)
print(dataset)
```

```python
dataset.to_dataframe()
```

```python
# Creating a DiskDataset
import tempfile
with tempfile.TemporaryDirectory() as data_dir:
  disk_dataset = dc.data.DiskDataset.from_numpy(X=X, y=y, data_dir=data_dir)
  print(disk_dataset)
```
